# üöó ELECTRIC VEHICLE SALES PREDICTION PROJECT

## üìå Problem Statement

**Objective:**
Predict and forecast Electric Vehicle (EV) sales trends based on historical data to help manufacturers, dealers, and policymakers make informed decisions.

## üéØ Goals:
1. Analyze EV adoption patterns across different regions
2. Identify key factors affecting EV sales (Model, Range, Price, Location)
3. Build a machine learning model to predict future EV sales
4. Understand consumer preferences for EV types and manufacturers

## üìä Dataset:
- **Source:** Washington State Department of Licensing
- **Records:** 135,000+ Electric Vehicles
- **Features:** 17 columns including Make, Model, Electric Range, Vehicle Type, Location, etc.

## üîß Approach:
- Data Cleaning & Preprocessing
- Exploratory Data Analysis (EDA)
- Feature Engineering
- Machine Learning Model (Regression/Time Series)
- Sales Forecasting

```python
print("üßπ DATA CLEANING & PREPROCESSING")
print("="*60)

print("\n1Ô∏è‚É£ Handling Missing Values...")

df['Electric Range'].fillna(df['Electric Range'].mean(), inplace=True)
df['Base MSRP'].fillna(0, inplace=True)
df['Legislative District'].fillna(df['Legislative District'].mode()[0], inplace=True)
df['Postal Code'].fillna(0, inplace=True)

df['County'].fillna('Unknown', inplace=True)
df['City'].fillna('Unknown', inplace=True)

print("‚úÖ Missing values handled!")

print("\n2Ô∏è‚É£ Removing Duplicates...")
before = len(df)
df = df.drop_duplicates()
after = len(df)
print(f"‚úÖ Removed {before - after} duplicate rows")

print("\n3Ô∏è‚É£ Final Check...")
print(f"üìä Clean Dataset Shape: {df.shape}")
print(f"‚ùå Remaining Missing Values: {df.isnull().sum().sum()}")

print("\n‚úÖ Data Cleaning Complete!")
```

```python
# EDA - Basic Analysis
print("üìä EXPLORATORY DATA ANALYSIS")
print("="*60)

# 1. Top EV Manufacturers
print("\nüöó Top 10 EV Manufacturers:")
print(df['Make'].value_counts().head(10))

# 2. Top EV Models
print("\nüî• Top 10 EV Models:")
print(df['Model'].value_counts().head(10))

# 3. EV Type Distribution
print("\n‚ö° Electric Vehicle Type Distribution:")
print(df['Electric Vehicle Type'].value_counts())

# 4. Year-wise EV Registration
print("\nüìÖ Model Year Distribution:")
print(df['Model Year'].value_counts().sort_index(ascending=False).head(10))

# 5. Top Cities with Most EVs
print("\nüèôÔ∏è Top 10 Cities with Most EVs:")
print(df['City'].value_counts().head(10))
```

```python
# Visualizations
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_palette("husl")

print("üìä Creating Visualizations...")

# 1. Top 10 Manufacturers
plt.figure(figsize=(12, 6))
df['Make'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 EV Manufacturers', fontsize=16, fontweight='bold')
plt.xlabel('Manufacturer', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. EV Type Distribution (Pie Chart)
plt.figure(figsize=(8, 8))
df['Electric Vehicle Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Electric Vehicle Type Distribution', fontsize=16, fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()

# 3. Year-wise EV Registration Trend
plt.figure(figsize=(14, 6))
year_counts = df['Model Year'].value_counts().sort_index()
year_counts = year_counts[year_counts.index >= 2010]  # 2010 se onwards
plt.plot(year_counts.index, year_counts.values, marker='o', linewidth=2, markersize=8, color='green')
plt.title('EV Registration Trend (2010 onwards)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Vehicles Registered', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("‚úÖ Visualizations Complete!")
```

```python
# More Visualizations

# 4. Electric Range Distribution
plt.figure(figsize=(12, 6))
df['Electric Range'].hist(bins=50, color='orange', edgecolor='black')
plt.title('Electric Range Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Electric Range (miles)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(df['Electric Range'].mean(), color='red', linestyle='--', linewidth=2, label=f"Mean: {df['Electric Range'].mean():.2f}")
plt.legend()
plt.tight_layout()
plt.show()

# 5. Top 10 Cities with EVs
plt.figure(figsize=(12, 6))
df['City'].value_counts().head(10).plot(kind='barh', color='teal')
plt.title('Top 10 Cities with Most EVs', fontsize=16, fontweight='bold')
plt.xlabel('Number of Vehicles', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.tight_layout()
plt.show()

# 6. Top 10 Models
plt.figure(figsize=(12, 6))
df['Model'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 EV Models', fontsize=16, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("‚úÖ All Visualizations Complete!")
```

### 5. **Vehicle Types:**
